### Classes definition
Generator and discriminator class. We also declare a function for weight intialization

```python
# Generator class
class Generator(nn.Module):
    def __init__(self, nz, ngf):
        super(Generator,self).__init__()

        self.lin = nn.Linear(nz, ngf*8*4*4)
        self.bn = nn.BatchNorm2d(ngf*8)
        self.tconv1 = nn.ConvTranspose2d(ngf*8, ngf*4, 4, 2, 1)
        self.bn1 = nn.BatchNorm2d(ngf*4)
        self.tconv2 = nn.ConvTranspose2d(ngf*4, ngf*4, 4, 2, 1)
        self.bn2 = nn.BatchNorm2d(ngf*4)
        self.tconv3 = nn.ConvTranspose2d(ngf*4, ngf*2, 4, 2, 1)
        self.bn3 = nn.BatchNorm2d(ngf*2)
        self.tconv4 = nn.ConvTranspose2d(ngf*2, ngf, 4, 2, 1)
        self.bn4 = nn.BatchNorm2d(ngf)
        self.tconv5 = nn.ConvTranspose2d(ngf, 3, 4, 2, 1)

        self.relu = nn.ReLU(True)
        self.tanh = nn.Tanh()
        self.ngf = ngf

    def forward(self, x):
        x = x.view(x.size(0), -1)
        x = self.lin(x)
        x = self.bn(x.view(x.size(0), self.ngf*8, 4, 4))
        x = self.relu(self.bn1(self.tconv1(x)))
        x = self.relu(self.bn2(self.tconv2(x)))
        x = self.relu(self.bn3(self.tconv3(x)))
        x = self.relu(self.bn4(self.tconv4(x)))
        x = self.tanh(self.tconv5(x))

        return x

# Discriminator class
class Discriminator (nn.Module):
    def __init__(self, nc, ndf):
        super(Discriminator, self).__init__()    
        self.conv1 = nn.Conv2d(nc, ndf, 4, 2, 1, bias=False)
        self.bn1 = nn.BatchNorm2d(ndf)
        self.conv2 = nn.Conv2d(ndf, ndf*2, 4, 2, 1, bias=False)
        self.bn2 = nn.BatchNorm2d(ndf*2)
        self.conv3 = nn.Conv2d(ndf*2, ndf*4, 4, 2, 1, bias=False)
        self.bn3 = nn.BatchNorm2d(ndf*4)
        self.conv4 = nn.Conv2d(ndf*4, ndf*4, 4, 2, 1, bias=False)
        self.bn4 = nn.BatchNorm2d(ndf*4)
        self.conv5 = nn.Conv2d(ndf*4, ndf*8, 4, 2, 1, bias=False)
        self.bn5 = nn.BatchNorm2d(ndf*8)
        self.lin = nn.Linear(ndf*8*4*4, 1)

        self.leakyrelu = nn.LeakyReLU(0.2, inplace=True)

    def forward (self, x):
        x = self.leakyrelu(self.bn1(self.conv1(x)))
        x = self.leakyrelu(self.bn2(self.conv2(x)))
        x = self.leakyrelu(self.bn3(self.conv3(x)))
        x = self.leakyrelu(self.bn4(self.conv4(x)))
        x = self.leakyrelu(self.bn5(self.conv5(x)))
        x = self.lin(x.view(x.size(0), -1))

        return x


# Weight initialization function
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('_bar') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)


# Summary of discriminator architecture
ndf = 32
nef = 64
model = Discriminator(3,ndf)
torchsummary.summary(model.to('cuda:0'),(3,128,128)) 

# Summary of generator architecture
model = Generator(100,ngf)
torchsummary.summary(model.to('cuda:0'),(100,1,1))
```

### Models and optimizers initialization

```python
# Define device as the available hardware
device = torch.device("cuda:0" if (torch.cuda.is_available()) else "cpu")

torch.manual_seed(4)

model_g = Generator(nz, ngf)
model_d = Discriminator(nc, ndf)

model_g.to(device)
model_d.to(device)

model_g.apply(weights_init)
model_d.apply(weights_init)
```

```python
lr_g = 2e-4
lr_d = 2e-4
weight_decay = 0
optimizer_D = torch.optim.Adam(model_d.parameters(), lr=lr_d, weight_decay=weight_decay, betas=(0.5,0.999))
optimizer_G = torch.optim.Adam(model_g.parameters(), lr=lr_g, weight_decay=weight_decay, betas=(0.5,0.999))
```

### Training

```python
batch = 128
nc = 3              # number of channels in training images
nz = 1024            # latent vector
ngf = 64           # Size of feature maps in generator
ndf = 32            # Size of feature maps in discriminator
epochs = 100


checkpoint=True
if checkpoint:
  checkpoint = torch.load('/content/drive/My Drive/models/lsgan.pt', map_location=lambda storage, loc: storage)
  model_d.load_state_dict(checkpoint['model_d'])
  model_g.load_state_dict(checkpoint['model_g'])
  cost_d = checkpoint['cost_d']
  cost_g = checkpoint['cost_g']
  epoch = checkpoint['epoch']


real_label = torch.Tensor(batch, 1).fill_(1.0)
fake_label = torch.Tensor(batch, 1).fill_(0.0)

it = len(cost_d)
cost_d=[]
cost_g=[]
img_list_noise = []
img_list_trunc = []
epoch = 0

for epoch in range(epoch, epoch+epochs):
    for i, img in enumerate(train_loader,1):
        it+=1
        real_imgs, _ = img

        #TRAIN D
        model_d.zero_grad()
        # Real batch
        real_score = D_real = model_d(real_imgs.to(device))
        loss_D_real = 0.5*torch.mean((D_real-real_label.to(device))**2)
        loss_D_real.backward()
        # Fake batch
        noise = torch.randn(batch, nz).to(device)
        fake_data = model_g(noise)
        fake_score = D_fake = model_d(fake_data.detach())
        loss_D_fake = 0.5*torch.mean((D_fake-fake_label.to(device))**2)
        loss_D_fake.backward()
        # Calculate D loss and update weights
        loss_D = loss_D_real  + loss_D_fake
        optimizer_D.step()

        #TRAIN G
        model_g.zero_grad()
        D_fake = model_d(fake_data)  
        loss_G = 0.5*torch.mean((D_fake-real_label.to(device))**2)
        loss_G.backward()
        optimizer_G.step()

        # Save losses
        cost_d.append(loss_D)
        cost_g.append(loss_G)

        # Print losses
        if i % 5 == 0:
            print('[%d/%d][%d/%d]\tLoss_D: %.4f\tLoss_G: %.4f'
                  % (epoch, epochs, i, len(train_loader), loss_D.item(), loss_G.item()))


    # Check how the generator is doing by saving G's output noise and 
    # truncated noise
    model_g.eval()
    trunc_noise = torch.FloatTensor(truncnorm.rvs(-1.5, 1.5, size=nz*batch))
    trunc_noise = trunc_noise.view(batch, nz)

    # Save generated images from noise
    with torch.no_grad():
        fake = model_g(trunc_noise.to(device)).detach().cpu()
    img_list_trunc.append(vutils.make_grid(fake, padding=2, normalize=True))
    vutils.save_image(img_list_trunc, '/content/drive/My Drive/Thesis/lsgan/imgs_trunc/lsgan_imgs_'+str(it)+'.jpg')

    # Save generated images from truncated noise
    with torch.no_grad():
        fake = model_g(noise.to(device)).detach().cpu()
    img_list_noise.append(vutils.make_grid(fake, padding=2, normalize=True))
    vutils.save_image(img_list_noise, '/content/drive/My Drive/Thesis/lsgan/imgs_noise/lsgan_imgs_'+str(it)+'.jpg')
    model_g.train()

    # Plot generator and discriminator loss
    plt.figure()
    plt.plot(cost_d, label='discrimiator')
    plt.plot(cost_d)
    plt.plot(cost_g, label='generator')
    plt.legend()

    # Save models and more info
    torch.save(
        {'model_d': model_d.state_dict(),
         'model_g': model_g.state_dict(),
         'epoch': epoch,
         'optimizer_d': optimizer_D.state_dict(),
         'optimizer_g': optimizer_G.state_dict(),
         'cost_d': cost_d,
         'cost_g': cost_g,
         'iter': it}, '/content/drive/My Drive/Thesis/lsgan/lsgan_github_'+str(it)+'.pt')
```
