Creation of class Autoencoder

```python
class AE_reconv(nn.Module):
    def __init__(self, nc, nef, ndf):
        #Encoder
        super(AE_reconv, self).__init__()
        self.nc = nc
        self.nef = nef
        self.ndf = ndf       
        self.conv1 = nn.Conv2d(nc,nef,3,2,1)
        self.bn1 = nn.BatchNorm2d(nef)        
        self.conv2 = nn.Conv2d(nef,nef*2,3,2,1)
        self.bn2 = nn.BatchNorm2d(nef*2)        
        self.conv3 = nn.Conv2d(nef*2,nef*2,3,2,1)
        self.bn3 = nn.BatchNorm2d(nef*2)       
        self.conv4 = nn.Conv2d(nef*2,nef*4,3,1,1)
        
        #Decoder        
        self.reconv1 = self.resize_conv(ndf*4,ndf*2,3)
        self.bn4 = nn.BatchNorm2d(ndf*2)    
        self.reconv2 = self.resize_conv(ndf*2,ndf*2,3)
        self.bn5 = nn.BatchNorm2d(ndf*2)      
        self.reconv3 = self.resize_conv(ndf*2,ndf,3)
        self.bn6 = nn.BatchNorm2d(ndf)        
        self.reconv4 = nn.Conv2d(ndf,nc,3,1,1)
        
        self.leakyrelu = nn.LeakyReLU()
        self.sigmoid = nn.Sigmoid()
        self.do = nn.Dropout(0.25)
        
    def encoder(self, x):
        #layer 1
        x = self.leakyrelu(self.bn1(self.conv1(x)))
        #layer 2
        x = self.leakyrelu(self.bn2(self.conv2(x)))
        #layer 3
        x = self.leakyrelu(self.bn3(self.conv3(x)))
        #layer 4
        x = self.leakyrelu(self.conv4(x))
        z = x.view(x.size(0),-1)
        
        return z
    
    
    def decoder(self,z):  
        #layer 5
        z = z.view(z.size(0), self.ndf*4, 16 ,16)
        z = self.leakyrelu(self.bn4(self.reconv1(z)))
        #layer 6
        z = self.leakyrelu(self.bn5(self.reconv2(z)))
        #layer 7
        z = self.leakyrelu(self.bn6(self.reconv3(z)))
        #Layer 8
        y = self.sigmoid(self.reconv4(z))
                 
        return y
    
    
    def forward (self,x):
        z = self.encoder(x)
        y = self.decoder(z)
        
        return y
  

        
    def resize_conv(self, i, o, k):
        
        reconv = nn.Sequential(nn.Upsample(scale_factor=2, mode='nearest'),
                            nn.ReflectionPad2d(1),
                            nn.Conv2d(i, o,
                            kernel_size=k, stride=1, padding=0),)
        return reconv
    
```

```python
class AE_tconv(nn.Module):
    def __init__(self, nc, nef, ndf):
        #Encoder
        super(AE_tconv, self).__init__()
        self.nc = nc
        self.nef = nef
        self.ndf = ndf       
        self.conv1 = nn.Conv2d(nc,nef,3,1,1)
        self.mp1 = nn.MaxPool2d(2, stride=2, return_indices=True)
        self.bn1 = nn.BatchNorm2d(nef)        
        self.conv2 = nn.Conv2d(nef,nef*2,3,1,1)
        self.mp2 = nn.MaxPool2d(2, stride=2, return_indices=True)
        self.bn2 = nn.BatchNorm2d(nef*2)        
        self.conv3 = nn.Conv2d(nef*2,nef*2,3,1,1)
        self.mp3 = nn.MaxPool2d(2,  stride=2, return_indices=True)
        self.bn3 = nn.BatchNorm2d(nef*2)       
        self.conv4 = nn.Conv2d(nef*2,nef*4,3,1,1)
        
        #Decoder        
        self.tconv1 = nn.ConvTranspose2d(ndf*4,ndf*2,3,1,1)
        self.mup1 = nn.MaxUnpool2d(2, stride=2)
        self.bn4 = nn.BatchNorm2d(ndf*2)    
        self.tconv2 = nn.ConvTranspose2d(ndf*2,ndf*2,3,1,1)
        self.mup2 = nn.MaxUnpool2d(2, stride=2)
        self.bn5 = nn.BatchNorm2d(ndf*2)      
        self.tconv3 = nn.ConvTranspose2d(ndf*2,ndf,3,1,1)
        self.mup3 = nn.MaxUnpool2d(2, stride=2)
        self.bn6 = nn.BatchNorm2d(ndf)        
        self.tconv4 = nn.ConvTranspose2d(ndf,nc,3,1,1)
        
        self.leakyrelu = nn.LeakyReLU()
        self.sigmoid = nn.Sigmoid()
        self.do = nn.Dropout(0.25)
        
    def encoder(self, x):
        #layer 1
        x, ind1 = self.mp1(self.conv1(x))
        x = self.leakyrelu(self.bn1(x))
        #layer 2
        x, ind2 = self.mp2(self.conv2(x))
        x = self.leakyrelu(self.bn2(x))
        #layer 3
        x, ind3 = self.mp3(self.conv3(x))
        x = self.leakyrelu(self.bn3(x))
        #layer 4
        x = self.leakyrelu(self.conv4(x))
        z = x.view(x.size(0),-1)
        
        return z, ind1, ind2, ind3
    
    
    def decoder(self,z, ind1, ind2, ind3):  
        #layer 5
        z = z.view(z.size(0), self.ndf*4, 16 ,16)
        z = self.leakyrelu(self.bn4(self.mup1(self.tconv1(z), ind3)))
        #layer 6
        z = self.leakyrelu(self.bn5(self.mup2(self.tconv2(z), ind2)))
        #layer 7
        z = self.leakyrelu(self.bn6(self.mup3(self.tconv3(z), ind1)))
        #Layer 8
        y = self.sigmoid(self.tconv4(z))
                 
        return y
    
    
    def forward (self,x):
        z, ind1, ind2, ind3 = self.encoder(x)
        y = self.decoder(z, ind1, ind2, ind3)
        
        return y
  

        
    def resize_conv(self, i, o, k):
        
        reconv = nn.Sequential(nn.Upsample(scale_factor=2, mode='nearest'),
                            nn.ReflectionPad2d(1),
                            nn.Conv2d(i, o,
                            kernel_size=k, stride=1, padding=0),)
        return reconv
    
```

### Model, criterion and optimizer initilization

```python
# ----------------------------------------------- DEFINE MODEL --------------------------------------------------------#
torch.manual_seed(42)
device = torch.device("cuda:0" if (torch.cuda.is_available()) else "cpu")
model = AE_tconv(3,32,32).to(device)
model.to(device)
# ------------------------------------------- CRITERION & OPTIMIZER ---------------------------------------------------#
lr = 1e-3
weight_decay = 1e-5
optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
```

```python
def mse_loss(input, target):
    r = input[:,0:1,:,:] - target[:,0:1,:,:]
    g = (input[:,1:2,:,:] - target[:,1:2,:,:])
    b = input[:,2:3,:,:] - target[:,2:3,:,:]
    
    r = torch.mean(r**2)
    g = torch.mean(g**2)
    b = torch.mean(b**2)
    
    mean = (r + g + b)/3
   
    return mean, r,g,b
```

### Training

```python
cost_list_train = []
cost_list_val = []
epoch = 0
```

```python
n_epochs = 20         #Number of epochs training
m = sys.maxsize       #Maximum value to check validation loss
epoch_since_min = 0   #Epochs since not reaching minimum loss in validation set
COST_TRAIN = 0        #training cost per epoch
lr_list = []


checkpoint = False
if checkpoint:
    checkpoint = torch.load('/content/drive/My Drive/models/ae_tconv_19.pth', map_location=lambda storage, loc: storage)
    model.load_state_dict(checkpoint['model_state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    epoch = checkpoint['epoch']
    cost_list_train = checkpoint['cost_list_train']
    cost_list_val = checkpoint['cost_list_val']
    lr_list = checkpoint['lr']
    model.cuda()
    
    
change_lr = False
if change_lr:
    optimizer.param_groups[0]['lr'] = 5e-4
    

for epoch in tqdm.tqdm(range(epoch,n_epochs+epoch)):
    COST_TRAIN = 0
    model.train()
    lr_list.append(lr)
    # ------------------------------------------------ TRAINING ----------------------------------------------------------#
    for i, data in enumerate(train_loader):
        img, _ = data        
        img = img.to(device)

        # Forward
        output = model(img)
        loss, r_loss, g_loss, b_loss = mse_loss(output, img)
        
        # Backprop
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        COST_TRAIN+=loss.item()
        
        # Print loss and show output images
        if i % 250 == 0:
            print(f"[{i}/{len(train_loader)}]")
            print('Losses: {:.4f} {:.4f} {:.4f} {:.4f}'.format(loss.data.item(), r_loss.data.item(), g_loss.data.item(), b_loss.data.item()))
            #print('Loss: {:.4f}'.format(loss.data.item()))
            
            #pic = to_img(output.to("cpu").data)
        
            fig = plt.figure(figsize=(16, 16))
        
            ax = plt.imshow(np.transpose(vutils.make_grid(output[:32], padding=2, normalize=True).cpu().detach().numpy(),(1,2,0)))
            ax.axes.get_xaxis().set_visible(False)
            ax.axes.get_yaxis().set_visible(False)
            plt.show(fig)
    # ------------------------------------------------ VALIDATION --------------------------------------------------------#        
    COST_VAL = 0 
    with torch.no_grad():
        print('Evaluating...')
        for j, data in enumerate(val_loader):
            img, _ = data        
            img = img.to(device)

            # Forward
            output = model(img)
            loss, r_loss, g_loss, b_loss = mse_loss(output, img)

            COST_VAL+=loss.item()

            if j % 50 == 0:
                print(f"[{j}/{len(val_loader)}]")
    
    cost_list_train.append(COST_TRAIN/len(train_loader))
    cost_list_val.append(COST_VAL/len(val_loader))

    # -------------------------------------------- EPOCHS SINCE MIN ------------------------------------------------------#
    print(m)
    if cost_list_val[epoch] < m:
        m = cost_list_val[epoch]
        epoch_since_min = 0
    else:
        epoch_since_min += 1

    print(cost_list_val)
    print(epoch_since_min)

    if epoch_since_min == 4:
        print('Training stopped')
        break

    
    # Save model, costs, epochs
    torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'cost_list_train': cost_list_train,
            'cost_list_val': cost_list_val,
            'lr': lr_list,
            }, 'model_tconv_'+str(epoch)+'.pth')
    
    # Plot cost graphs
    plt.figure()
    plt.plot(cost_list_train, label='training')
    plt.title('TRAIN COST')
    plt.plot(cost_list_val, label='evaluation')
    plt.title('VAL COST')
    plt.legend()
    print(epoch)
```

### Analysis 
Let's load AE model with resize convolutions and transpose convolutions 

```python
model_tconv = AE_tconv(3,32,32).to(device)
model_reconv = AE_reconv(3,32,32).to(device)

# Load tconv model
checkpoint = torch.load('/content/drive/My Drive/models/ae_tconv_19.pth', map_location=lambda storage, loc: storage)
model_tconv.load_state_dict(checkpoint['model_state_dict'])
# Load reconv model
checkpoint = torch.load('/content/drive/My Drive/models/ae_reconv_19.pth', map_location=lambda storage, loc: storage)
model_reconv.load_state_dict(checkpoint['model_state_dict'])
```

```python
# ------------------------------------------------ ANALYSIS ----------------------------------------------------------#
model_tconv.eval()
model_reconv.eval()
analysis_loader = iter(test_dataset)

fig = plt.figure(figsize=(20,10))

for i in range(5):
    x, _ = next(analysis_loader)
    with torch.no_grad():
        x = x.to(device)
        output_tconv = model_tconv(x.unsqueeze(0)).squeeze().cpu().numpy()
        output_reconv = model_reconv(x.unsqueeze(0)).squeeze().cpu().numpy()
        
        #Plot original image
        plt.subplot(3,5,i+1)
        x = x.permute(1,2,0)
        plt.imshow(x.cpu().numpy())
        plt.axis('off')

        #Plot output image
        plt.subplot(3,5,i+5+1)
        plt.imshow(np.transpose(output_tconv,(1,2,0)))
        plt.axis('off')
        
        #Plot output image
        plt.subplot(3,5,i+10+1)
        plt.imshow(np.transpose(output_reconv,(1,2,0)))
        plt.axis('off')

plt.show(fig)
```

We see the in the first row the original images input to the AE. In the second row, the output images from theAE with transpose convolutions and maxpooling. And, in the last one the AE with resize convolution. We see how tranposed convolutions and maxpooling causes what it is known as "checkerboard patterns". Therefore, we have a higher quality image reconstruction with resize convolutions.
