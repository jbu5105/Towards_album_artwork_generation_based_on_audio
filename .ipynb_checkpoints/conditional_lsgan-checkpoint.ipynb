And let's create our custom dataset class to load album covers and audio features at the same iteration

```python
class concatDataset(Dataset):
    def __init__(self, audio, cover):
        super(concatDataset, self).__init__()
        self.dataset_audio = audio
        self.dataset_cover = cover

    def __getitem__(self, index):
        audio = self.dataset_audio[index]
        cover = self.dataset_cover[index]
        return audio, cover

    def __len__(self):
        return len(self.dataset_audio)


def npy_loader(npy):
    sample = torch.from_numpy(np.load(npy))
    return sample
```

Now, we define the DataLoader we will use during training

```python
batch = 64

#Cover
dataroot_cover = 'multimodal_dataset/audio-cover_dataset/cover/train'
train_dataset_cover = dset.ImageFolder(root=dataroot_cover,
                           transform=transforms.Compose([
                               transforms.ToTensor(),
                               transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
                           ]))


#Audio
dataroot_audio = 'multimodal_dataset/audio-cover_dataset/audio/train'
train_dataset_audio = dset.DatasetFolder(root=dataroot_audio, loader=npy_loader
                                         , extensions='.npy')

#DataLoader
train_loader = torch.utils.data.DataLoader(dataset=concatDataset(train_dataset_audio, train_dataset_cover),
                                           batch_size=batch, shuffle=True, drop_last=True)
```

### Classes definition
Generator and discriminator class. We also declare a function for weight initialization

```python
class generator(nn.Module):
    def __init__(self, nz, ngf, i_mlp=1160, o_mlp=512):
        super(generator,self).__init__()

        self.lin = nn.Linear(nz, ngf*8*4*4)
        self.bn = nn.BatchNorm2d(ngf*8, affine=False)
        self.tconv1 = nn.ConvTranspose2d(ngf*8, ngf*4, 4, 2, 1)
        self.bn1 = nn.BatchNorm2d(ngf*4, affine=False)
        self.tconv2 = nn.ConvTranspose2d(ngf*4, ngf*4, 4, 2, 1)
        self.bn2 = nn.BatchNorm2d(ngf*4, affine=False)
        self.tconv3 = nn.ConvTranspose2d(ngf*4, ngf*2, 4, 2, 1)
        self.bn3 = nn.BatchNorm2d(ngf*2, affine=False)
        self.tconv4 = nn.ConvTranspose2d(ngf*2, ngf, 4, 2, 1)
        self.bn4 = nn.BatchNorm2d(ngf, affine=False)
        self.tconv5 = nn.ConvTranspose2d(ngf, 3, 4, 2, 1)

        # MLP
        self.mlp = nn.Sequential(nn.Linear(i_mlp, i_mlp - abs(i_mlp - o_mlp)//3),
                                nn.BatchNorm1d(i_mlp - abs(i_mlp - o_mlp)//3),
                                nn.LeakyReLU(0.2, inplace=True),
                                nn.Linear(i_mlp - abs(i_mlp - o_mlp)//3, i_mlp 
                                          - (abs(i_mlp - o_mlp)//3)*2),
                                nn.BatchNorm1d(i_mlp - (abs(i_mlp - o_mlp)//3)*2),
                                nn.LeakyReLU(0.2, inplace=True),
                                nn.Linear(i_mlp - (abs(i_mlp - o_mlp)//3)*2, o_mlp),
                                nn.BatchNorm1d(o_mlp))
      

        # Activation - Layer 1 - Gamma
        self.a1_gamma = nn.Sequential(nn.Linear(o_mlp, (o_mlp + ngf*8) // 2),
                                      nn.BatchNorm1d((o_mlp + ngf*8) // 2),
                                      nn.LeakyReLU(0.2, inplace=True),
                                      nn.Linear((o_mlp + ngf*8) // 2, ngf*8),
                                      nn.BatchNorm1d(ngf*8))


        # Activation - Layer 2 - Gamma
        self.a2_gamma = nn.Sequential(nn.Linear(o_mlp, (o_mlp + ngf*4) // 2),
                                      nn.BatchNorm1d((o_mlp + ngf*4) // 2),
                                      nn.LeakyReLU(0.2, inplace=True),
                                      nn.Linear((o_mlp + ngf*4) // 2, ngf*4),
                                      nn.BatchNorm1d(ngf*4))

        # Activation - Layer 3 - Gamma
        self.a3_gamma = nn.Sequential(nn.Linear(o_mlp, (o_mlp + ngf*4) // 2),
                                      nn.BatchNorm1d((o_mlp + ngf*4) // 2),
                                      nn.LeakyReLU(0.2, inplace=True),
                                      nn.Linear((o_mlp + ngf*4) // 2, ngf*4),
                                      nn.BatchNorm1d(ngf*4))

        # Activation - Layer 4 - Gamma
        self.a4_gamma = nn.Sequential(nn.Linear(o_mlp, (o_mlp + ngf*2) // 2),
                                      nn.BatchNorm1d((o_mlp + ngf*2) // 2),
                                      nn.LeakyReLU(0.2, inplace=True),
                                      nn.Linear((o_mlp + ngf*2) // 2, ngf*2),
                                      nn.BatchNorm1d(ngf*2))

        # Activation - Layer 5 - Gamma
        self.a5_gamma = nn.Sequential(nn.Linear(o_mlp, (o_mlp + ngf) // 2),
                                      nn.BatchNorm1d((o_mlp + ngf) // 2),
                                      nn.LeakyReLU(0.2, inplace=True),
                                      nn.Linear((o_mlp + ngf) // 2, ngf),
                                      nn.BatchNorm1d(ngf))


        # Activation - Layer 1 - Beta
        self.a1_beta = nn.Sequential(nn.Linear(o_mlp, (o_mlp + ngf*8) // 2),
                                      nn.BatchNorm1d((o_mlp + ngf*8) // 2),
                                      nn.LeakyReLU(0.2, inplace=True),
                                      nn.Linear((o_mlp + ngf*8) // 2, ngf*8),
                                      nn.BatchNorm1d(ngf*8))

        # Activation - Layer 2 - Beta
        self.a2_beta = nn.Sequential(nn.Linear(o_mlp, (o_mlp + ngf*4) // 2),
                                      nn.BatchNorm1d((o_mlp + ngf*4) // 2),
                                      nn.LeakyReLU(0.2, inplace=True),
                                      nn.Linear((o_mlp + ngf*4) // 2, ngf*4),
                                      nn.BatchNorm1d(ngf*4))

        # Activation - Layer 3 - Beta
        self.a3_beta = nn.Sequential(nn.Linear(o_mlp, (o_mlp + ngf*4) // 2),
                                      nn.BatchNorm1d((o_mlp + ngf*4) // 2),
                                      nn.LeakyReLU(0.2, inplace=True),
                                      nn.Linear((o_mlp + ngf*4) // 2, ngf*4),
                                      nn.BatchNorm1d(ngf*4))

        # Activation - Layer 4 - Beta
        self.a4_beta = nn.Sequential(nn.Linear(o_mlp, (o_mlp + ngf*2) // 2),
                                      nn.BatchNorm1d((o_mlp + ngf*2) // 2),
                                      nn.LeakyReLU(0.2, inplace=True),
                                      nn.Linear((o_mlp + ngf*2) // 2, ngf*2),
                                      nn.BatchNorm1d(ngf*2))

        # Activation - Layer 5 - Beta
        self.a5_beta = nn.Sequential(nn.Linear(o_mlp, (o_mlp + ngf) // 2),
                                      nn.BatchNorm1d((o_mlp + ngf) // 2),
                                      nn.LeakyReLU(0.2, inplace=True),
                                      nn.Linear((o_mlp + ngf) // 2, ngf),
                                      nn.BatchNorm1d(ngf))


        self.relu = nn.ReLU(True)
        self.tanh = nn.Tanh()
        self.ngf = ngf

    def forward(self, noise, x_audio):
        # MLP
        x_audio = self.mlp(x_audio)

        #Gamma
        gamma1 = self.a1_gamma(x_audio).view(x_audio.size(0), -1, 1, 1)
        gamma2 = self.a2_gamma(x_audio).view(x_audio.size(0), -1, 1, 1)
        gamma3 = self.a3_gamma(x_audio).view(x_audio.size(0), -1, 1, 1)
        gamma4 = self.a4_gamma(x_audio).view(x_audio.size(0), -1, 1, 1)
        gamma5 = self.a5_gamma(x_audio).view(x_audio.size(0), -1, 1, 1)

        #Beta
        beta1 = self.a1_beta(x_audio).view(x_audio.size(0), -1, 1, 1)
        beta2 = self.a2_beta(x_audio).view(x_audio.size(0), -1, 1, 1)
        beta3 = self.a3_beta(x_audio).view(x_audio.size(0), -1, 1, 1)
        beta4 = self.a4_beta(x_audio).view(x_audio.size(0), -1, 1, 1)
        beta5 = self.a5_beta(x_audio).view(x_audio.size(0), -1, 1, 1)
        
        #Convolutional net
        x = noise.view(noise.size(0), -1)
        x = self.lin(x)
        x = self.relu(beta1*self.bn(x.view(x.size(0), self.ngf*8, 4, 4))+gamma1)
        x = self.relu(beta2*self.bn1(self.tconv1(x))+gamma2)
        x = self.relu(beta3*self.bn2(self.tconv2(x))+gamma3)
        x = self.relu(beta4*self.bn3(self.tconv3(x))+gamma4)
        x = self.relu(beta5*self.bn4(self.tconv4(x))+gamma5)
        x = self.tanh(self.tconv5(x))

        return x


    def init_gamma_beta(self):
      # Gamma
      nn.init.normal_(self.a1_gamma._modules['4'].weight.data, 1, 0.02)
      nn.init.normal_(self.a2_gamma._modules['4'].weight.data, 1, 0.02)
      nn.init.normal_(self.a3_gamma._modules['4'].weight.data, 1, 0.02)
      nn.init.normal_(self.a4_gamma._modules['4'].weight.data, 1, 0.02)
      nn.init.normal_(self.a5_gamma._modules['4'].weight.data, 1, 0.02)
      # Beta
      nn.init.normal_(self.a1_beta._modules['4'].weight.data, 0, 0.02)
      nn.init.normal_(self.a2_beta._modules['4'].weight.data, 0, 0.02)
      nn.init.normal_(self.a3_beta._modules['4'].weight.data, 0, 0.02)
      nn.init.normal_(self.a4_beta._modules['4'].weight.data, 0, 0.02)
      nn.init.normal_(self.a5_beta._modules['4'].weight.data, 0, 0.02)


class discriminator (nn.Module):
    def __init__(self, nc, ndf):
        super(discriminator, self).__init__()    
        self.conv1 = nn.Conv2d(nc, ndf, 4, 2, 1, bias=False)
        self.bn1 = nn.BatchNorm2d(ndf)
        self.conv2 = nn.Conv2d(ndf, ndf*2, 4, 2, 1, bias=False)
        self.bn2 = nn.BatchNorm2d(ndf*2)
        self.conv3 = nn.Conv2d(ndf*2, ndf*4, 4, 2, 1, bias=False)
        self.bn3 = nn.BatchNorm2d(ndf*4)
        self.conv4 = nn.Conv2d(ndf*4, ndf*4, 4, 2, 1, bias=False)
        self.bn4 = nn.BatchNorm2d(ndf*4)
        self.conv5 = nn.Conv2d(ndf*4, ndf*8, 4, 2, 1, bias=False)
        self.bn5 = nn.BatchNorm2d(ndf*8)
        self.lin = nn.Linear(ndf*8*4*4, 1)

        self.leakyrelu = nn.LeakyReLU(0.2, inplace=True)

    def forward (self, x):
        x = self.leakyrelu(self.bn1(self.conv1(x)))
        x = self.leakyrelu(self.bn2(self.conv2(x)))
        x = self.leakyrelu(self.bn3(self.conv3(x)))
        x = self.leakyrelu(self.bn4(self.conv4(x)))
        x = self.leakyrelu(self.bn5(self.conv5(x)))
        x = self.lin(x.view(x.size(0), -1))

        return x



def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        m.weight.data.normal_(0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
      if hasattr(m.weight, 'data') and hasattr(m.bias, 'data'):
        nn.init.normal_(m.weight.data, 0.0, 1)
        nn.init.normal_(m.bias.data, 0.0, 1)
    if classname.find('Linear')!= -1:
        stdv = 1. / math.sqrt(m.weight.size(1))
        m.weight.data.uniform_(-stdv, stdv)
        if m.bias is not None:
            m.bias.data.uniform_(-stdv, stdv)
```

### Models and optimizers initialization

We initialize the conditioned generator and discriminator, apply weight initialization and initialize FiLM layer to zero mean and unit variance.

```python
# Define device as the available hardware
device = torch.device("cuda:0" if (torch.cuda.is_available()) else "cpu")

batch = 128
nc = 3              # number of channels in training images
nz = 1024            # latent vector
ngf = 64           # Size of feature maps in generator
ndf = 64            # Size of feature maps in discriminator

model_g = generator(nz, ngf)
model_d = discriminator(nc, ndf)

model_g.to(device)
model_d.to(device)

model_g.apply(weights_init)
model_d.apply(weights_init)
# Init FiLM layers to 0 mean and 1 variance
model_g.init_gamma_beta()
```

```python
lr_g = 2e-4
lr_d = 2e-4
weight_decay = 0
optimizer_D = torch.optim.Adam(model_d.parameters(), lr=lr_d, weight_decay=weight_decay, betas=(0.5,0.999))
optimizer_G = torch.optim.Adam(model_g.parameters(), lr=lr_g, weight_decay=weight_decay, betas=(0.5,0.999))
```

### Training

```python
epochs = 100
cost_d=[]
cost_g=[]
img_list_noise=[]
img_list_trunc=[]
epoch = 0

checkpoint=True
if checkpoint:
  checkpoint = torch.load('/content/drive/My Drive/models/clsgan.pt', map_location=lambda storage, loc: storage)
  model_d.load_state_dict(checkpoint['model_d'])
  model_g.load_state_dict(checkpoint['model_g'])
  cost_d = checkpoint['cost_d']
  cost_g = checkpoint['cost_g']
  epoch = checkpoint['epoch']
  it = len(cost_d)


real_label = torch.Tensor(batch, 1).fill_(1.0)
fake_label = torch.Tensor(batch, 1).fill_(0.0)


model_g.train()
model_d.train()
for epoch in range(epoch, epoch+epochs):
    epoch = epoch + 1
    for i, data in enumerate(train_loader,1):
        it+=1
        audio, cover = data
        audio_x, audio_y = audio
        cover_x, cover_y = cover

        audio_x = audio_x.float().to(device)

        #Train D
        model_d.zero_grad()
        # Real batch
        real_score = D_real = model_d(cover_x.to(device))
        loss_D_real = 0.5*torch.mean((D_real-real_label.to(device))**2)
        loss_D_real.backward()
        # Fake batch
        noise = torch.randn(batch, nz).to(device)
        fake_data = model_g(noise, audio_x)
        fake_score = D_fake = model_d(fake_data.detach())
        loss_D_fake = 0.5*torch.mean((D_fake-fake_label.to(device))**2)
        loss_D_fake.backward()
        # Calculate D loss and update weights
        loss_D = loss_D_real  + loss_D_fake
        if it % 2 == 0:
          optimizer_D.step()

        #Train G
        model_g.zero_grad()
        D_fake = model_d(fake_data)
        loss_G = 0.5*torch.mean((D_fake-real_label.to(device))**2)
        loss_G.backward()
        optimizer_G.step()

        # Save losses
        cost_d.append(loss_D)
        cost_g.append(loss_G)

        # Print losses
        if i % 5 == 0:
            print('[%d/%d][%d/%d]\tLoss_D: %.4f\tLoss_G: %.4f'
                  % (epoch, epochs, i, len(train_loader), loss_D.item(), loss_G.item()))


  
    # Check how the generator is doing by saving G's output noise and 
    # truncated noise  
    model_g.eval()
    trunc_noise = torch.FloatTensor(truncnorm.rvs(-2, 2, size=nz*batch))
    trunc_noise = trunc_noise.view(batch, nz)

    # Save generated images from noise
    with torch.no_grad():
        fake = model_g(noise.to(device), audio_x).detach().cpu()
    img_list_noise.append(vutils.make_grid(fake, padding=2, normalize=True))
    vutils.save_image(img_list_noise, '/content/drive/My Drive/Thesis/conditional_lsgan/imgs_noise/imgs_epoch_'+str(epoch)+'_'+str(it)+'.jpg')

    # Save generated images from truncated noise
    with torch.no_grad():
        fake = model_g(trunc_noise.to(device), audio_x).detach().cpu()
    img_list_trunc.append(vutils.make_grid(fake, padding=2, normalize=True))
    vutils.save_image(img_list_trunc, '/content/drive/My Drive/Thesis/conditional_lsgan/imgs_trunc/imgs_epoch_'+str(epoch)+'_'+str(it)+'.jpg')

    # Save genre from audio features used
    with open('/content/drive/My Drive/Thesis/conditional_lsgan/genre/genre_epoch_'+str(epoch)+'_'+str(it)+'.npy', 'wb') as f:
      np.save(f, audio_y)
    model_g.train()

    # Plot generator and discriminator loss
    plt.figure()
    plt.plot(cost_d, label='discrimiator')
    plt.plot(cost_d)
    plt.plot(cost_g, label='generator')
    plt.legend()

    # Save models and more info
    torch.save(
        {'model_d': model_d.state_dict(),
         'model_g': model_g.state_dict(),
         'epoch': epoch,
         'optimizer_d': optimizer_D.state_dict(),
         'optimizer_g': optimizer_G.state_dict(),
         'cost_d': cost_d,
         'cost_g': cost_g,
         'iter': it}, '/content/drive/My Drive/Thesis/conditional_lsgan/lsgan_epoch_'+str(epoch)+'_'+str(it)+'.pt')
```

#### Check generated images by the genres of audio features
We are going to make an experiment where we generate album covers from the same random vector for different genres. We then analyze the outputs.

We are going to first save in a dictionary how music genres are coded by pytorch DataLoader

```python
import os

dict_genres = {}
count = 0
last_genre = ''
for root, dirs, files in sorted(os.walk('multimodal_dataset/audio-cover_dataset/cover/train')):
    for file in files:
      genre = root.replace(os.sep, '/').split('/')[3]
      if genre != last_genre:  
        count+=1
        dict_genres[count-1]=genre
      last_genre = root.replace(os.sep, '/').split('/')[3]
```

```python
dict_genres
```

We define a new dataloader

```python
batch = 64
#Cover
dataroot_cover = 'multimodal_dataset/audio-cover_dataset/cover/train'
train_dataset_cover = dset.ImageFolder(root=dataroot_cover,
                           transform=transforms.Compose([
                               transforms.ToTensor(),
                               transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
                           ]))


#Audio
dataroot_audio = 'multimodal_dataset/audio-cover_dataset/audio/train'
train_dataset_audio = dset.DatasetFolder(root=dataroot_audio, loader=npy_loader
                                         , extensions='.npy')

train_loader = torch.utils.data.DataLoader(dataset=concatDataset(train_dataset_audio, train_dataset_cover),
                                           batch_size=batch, shuffle=True, drop_last=True)
```

We are going to analyze the outputs for 8 genres: blues, electronic, jazz, metal, pop, punk, rap and rock. 

```python
genres_test = []
for i in dict_genres:
  if dict_genres[i] == 'blues':
    genres_test.append(i)
  elif dict_genres[i] == 'electronic':
    genres_test.append(i)
  elif dict_genres[i] == 'jazz':
    genres_test.append(i)
  elif dict_genres[i] == 'metal':
    genres_test.append(i)
  elif dict_genres[i] == 'pop':
    genres_test.append(i)
  elif dict_genres[i] == 'punk':
    genres_test.append(i)
  elif dict_genres[i] == 'rap':
    genres_test.append(i)
  elif dict_genres[i] == 'rock':
    genres_test.append(i)
```

And we are now going to run our 8 random vector for each of the 8 genres. Therefore as output we have that each of the columns corresponds to the same noise vector input
into the generator, and each of the rows corresponds to different audio feature vectors
from the same genre. See analysis of output in the document

```python
device = torch.device("cuda:0" if (torch.cuda.is_available()) else "cpu")

model_g.eval()
model_g.to(device)

test_batch = torch.zeros(1,3,128,128)
test_genre = torch.zeros(1).unsqueeze(0)

noise = torch.randn(8, 1024, 1, 1)

# Iterate for each of the selected genres
for genre_count, i in enumerate(genres_test):
  count = 0
  test_audio = torch.zeros(1, 1160)
  # Load data from train loader
  for data in train_loader:
    audio, cover = data
    audio_x, audio_y = audio
    cover_x, cover_y = cover

    # Load each of the audio features independtly
    for num, j in enumerate(audio_y):
      # Concatenate the audio features for genre i
      if j == i and count<8:
        test_audio = torch.cat((test_audio, audio_x[num].unsqueeze(0)), 0)
        test_genre = torch.cat((test_genre, audio_y[num].unsqueeze(0).unsqueeze(0)), 0)
        count+=1
    # We run the model for the 8 random vector and the 8 audio features
    # gathered and concatenate them
    if count == 8:
      test_audio = test_audio[1:]
      fake = model_g(noise.to(device), test_audio.float().to(device)).detach().cpu()
      test_batch = torch.cat((test_batch, fake), 0)
      break 

test_batch = test_batch[1:]
test_genre = test_genre[1:]
```
